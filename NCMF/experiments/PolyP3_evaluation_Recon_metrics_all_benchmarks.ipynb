Here, we compare the reconstructed matrices from NCMF, NIMCGCN, GRGMF and CSGNN over the test split.

```python
import pandas as pd
import numpy as np
```

```python
sample_id = 3
```

```python
ncmf_recon_file = f"../../datasets/NCMF/PolyP3/{sample_id}/X1.npy"
ncmf_x1 = np.load(ncmf_recon_file, allow_pickle=True)
print(ncmf_x1.shape)
ncmf_x1
```

```python
grgmf_recon_file = f"../../GRGMF/output/PolyP3_{sample_id}_predicted_matrix.xlsx"
grgmf_x1 = pd.read_excel(grgmf_recon_file, engine="openpyxl", index_col=0)
print(grgmf_x1.shape)
grgmf_x1
```

```python
nimcgcn_drug_repr = np.load(f"../../NIMCGCN/data/PolyP3/{sample_id}/drug_repr.npy", allow_pickle=True)
nimcgcn_protein_repr = np.load(f"../../NIMCGCN/data/PolyP3/{sample_id}/protein_repr.npy", allow_pickle=True)
print(nimcgcn_drug_repr.shape, nimcgcn_protein_repr.shape)
```

```python
nimcgcn_x1 = np.matmul(nimcgcn_drug_repr, np.transpose(nimcgcn_protein_repr))
print(nimcgcn_x1.shape)
nimcgcn_x1
```

```python
original_x1 = pd.read_csv(f"../../datasets/NCMF/PolyP3/drug-protein.csv", header=None)
print(original_x1.shape)
original_x1
```

```python
test_x1 = pd.read_csv(f"../../datasets/NCMF/PolyP3/sampled{sample_id}_link.dat.test", sep="\t", header=None)
test_x1.columns = ["left", "right", "value"]
test_x1
```

```python
def get_value(row, array):
    i = int(row["left"])
    j = int(row["right"]) - 645
    return array[i][j]
```

```python
test_x1["ncmf_recon"] = test_x1.apply(lambda row: get_value(row, ncmf_x1), axis = 1)
test_x1
```

```python
test_x1["grgmf_recon"] = test_x1.apply(lambda row: get_value(row, grgmf_x1.values), axis = 1)
test_x1
```

```python
test_x1["nimcgcn_recon"] = test_x1.apply(lambda row: get_value(row, nimcgcn_x1), axis = 1)
test_x1
```

```python
from sklearn.preprocessing import MinMaxScaler
test_x1["ncmf_recon_min_max_scaled"] = MinMaxScaler().fit_transform(test_x1[["ncmf_recon"]])
```

```python
test_x1
```

```python
y_true = list(test_x1["value"])
```

```python
from sklearn.metrics import mean_squared_error
print("NCMF")
print(mean_squared_error(y_true, list(test_x1["ncmf_recon"])))
print("GRGMF")
print(mean_squared_error(y_true, list(test_x1["grgmf_recon"])))
print("NIMCGCN")
print(mean_squared_error(y_true, list(test_x1["nimcgcn_recon"])))
```

```python
from sklearn.metrics import roc_auc_score, precision_recall_curve, auc
auroc = roc_auc_score(y_true, list(test_x1["ncmf_recon_min_max_scaled"]))
precision, recall, thresholds = precision_recall_curve(y_true, list(test_x1["ncmf_recon_min_max_scaled"]))
auprc = auc(recall, precision)
print("NCMF")
print(f"AUROC = {auroc}")
print(f"AUPRC = {auprc}")
```

```python
auroc = roc_auc_score(y_true, list(test_x1["ncmf_recon"]))
precision, recall, thresholds = precision_recall_curve(y_true, list(test_x1["ncmf_recon"]))
auprc = auc(recall, precision)
print("NCMF")
print(f"AUROC = {auroc}")
print(f"AUPRC = {auprc}")
```

```python
auroc = roc_auc_score(y_true, list(test_x1["grgmf_recon"]))
precision, recall, thresholds = precision_recall_curve(y_true, list(test_x1["grgmf_recon"]))
auprc = auc(recall, precision)
print("GRGMF")
print(f"AUROC = {auroc}")
print(f"AUPRC = {auprc}")
```

```python
auroc = roc_auc_score(y_true, list(test_x1["nimcgcn_recon"]))
precision, recall, thresholds = precision_recall_curve(y_true, list(test_x1["nimcgcn_recon"]))
auprc = auc(recall, precision)
print("NIMCGCN")
print(f"AUROC = {auroc}")
print(f"AUPRC = {auprc}")
```
