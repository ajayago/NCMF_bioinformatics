This notebook takes in as input the raw files for cell lines obtained from the CCLE website (build 21Q3) - https://depmap.org/portal/download/all/

* mutations - CCLE_mutations.csv
* gene expression - CCLE_expression.csv
* audrc scores - sanger_dose_response.csv (search for this file in the search bar of the above link)


```python
import pandas as pd
import numpy as np
```

```python
data_dir = "../../datasets/NCMF/CellLine/"
```

```python
gene_exp = pd.read_csv(data_dir + "CCLE_expression.csv", index_col = 0)
print(gene_exp.shape)
gene_exp
```

```python
mut = pd.read_csv(data_dir + "CCLE_mutations.csv")
mut
```

```python
mut.Variant_Type.value_counts()
```

```python
mut.Variant_Classification.value_counts()
```

```python
mut.Variant_annotation.value_counts()
```

```python
cnv = pd.read_csv(data_dir + "CCLE_cn_segments_annotated_21Q3.txt", sep = "\t")
cnv
```

```python
audrc = pd.read_csv(data_dir + "sanger-dose-response.csv")
audrc
```

```python
overlapping_ids = set(gene_exp.index) & set(mut.DepMap_ID) & set(cnv["DepMap-ID"]) & set(audrc.ARXSPAN_ID)
len(overlapping_ids)
```

#### Mutation processing

```python
mut_reduced = mut[mut.DepMap_ID.isin(overlapping_ids)]
```

```python
mutation_filtered = mut_reduced[['Hugo_Symbol', 'Variant_annotation','DepMap_ID']]

mutations_needed = ['damaging', 'other non-conserving']
mutation_filtered = mutation_filtered[mutation_filtered['Variant_annotation'].isin(mutations_needed)]

# binary file
mutation_filtered['mutated'] = 1
mutation_filtered_1 = pd.pivot_table(mutation_filtered, 'mutated', index=['DepMap_ID'], columns=['Hugo_Symbol'], aggfunc = 'first')
mutation_filtered_1.fillna(0,inplace=True)
```

```python
mutation_filtered_1
```

#### CNV processing

```python
cnv_reduced = cnv[cnv["DepMap-ID"].isin(overlapping_ids)]
```

```python
cnv_filtered = pd.pivot_table(cnv_reduced, 'status', index=['DepMap-ID'], columns=['genesymbol'], aggfunc = 'first')
```

```python
cnv_filtered1 = cnv_filtered.replace({'U':0, '+':1, '-':-1})
```

```python
cnv_filtered1
```

#### Gene expression processing

```python
gene_exp_reduced = gene_exp[gene_exp.index.isin(overlapping_ids)]
gene_expression_1 = gene_exp_reduced.rename(columns=lambda x: x.split('(')[0].strip())
```

```python
gene_expression_1
```

```python
overlapping_genes = list(set(mutation_filtered_1.columns) & set(cnv_filtered1.columns) & set(gene_expression_1.columns))
overlapping_genes = sorted(overlapping_genes)
len(overlapping_genes)
```

```python
mut_csv = mutation_filtered_1[overlapping_genes]
mut_csv = mut_csv.sort_index()
mut_csv
```

```python
cnv_csv = cnv_filtered1[overlapping_genes]
cnv_csv = cnv_csv.sort_index()
cnv_csv
```

```python
gene_exp_csv = gene_expression_1[overlapping_genes]
gene_exp_csv = gene_exp_csv.sort_index()
gene_exp_csv
```

```python
print("Mean", np.mean(gene_exp_csv.values))
print("Min", np.min(gene_exp_csv.values))
print("Max", np.max(gene_exp_csv.values))
print("Median", np.median(gene_exp_csv.values))
```

```python
# scaling gene expression
from sklearn.preprocessing import StandardScaler
gene_exp_scaler = StandardScaler()
gene_exp_scaled = gene_exp_scaler.fit_transform(gene_exp_csv.values)
print("Mean", np.mean(gene_exp_scaled))
print("Min", np.min(gene_exp_scaled))
print("Max", np.max(gene_exp_scaled))
print("Median", np.median(gene_exp_scaled))
```

```python
gene_exp_scaled_csv = pd.DataFrame(gene_exp_scaled)
gene_exp_scaled_csv.columns = list(gene_exp_csv.columns)
gene_exp_scaled_csv.index = list(gene_exp_csv.index)
gene_exp_scaled_csv
```

#### AUDRC Processing

```python
audrc_reduced = audrc[audrc["ARXSPAN_ID"].isin(overlapping_ids)]
```

```python
audrc_reduced['DRUG_NAME_new'] = audrc_reduced['DRUG_NAME'].str.split(',').str[0]
```

```python
#take the latest drug auc value
cellline_drug_auc = pd.pivot_table(audrc_reduced, 'AUC_PUBLISHED', index=['ARXSPAN_ID'], columns=['DRUG_NAME_new'], aggfunc = 'last')
```

```python
cellline_drug_auc
```

```python
np.where(cellline_drug_auc.values == 1)
```

##### Imputing NaN values with KNN


```python
from sklearn.impute import KNNImputer
```

```python
# combining mutations, cnv, unscaled gene exp (17515 genes), raw audrc
knn_input_matrix = np.concatenate((mut_csv.values, cnv_csv.values, gene_exp_csv.values, cellline_drug_auc.values), axis = 1)
knn_input_matrix.shape
```

```python
imputer = KNNImputer(n_neighbors=5)
imputer.fit(knn_input_matrix) # fit on combined matrix of mutations, cnv, unscaled gene exp and raw audrc
```

```python
imputed_np = imputer.transform(knn_input_matrix)
imputed_np.shape
```

```python
imputed_np[:, -396:].shape
```

```python
cellline_drug_auc_imputed = pd.DataFrame(imputed_np[:, -396:])
cellline_drug_auc_imputed.columns = list(cellline_drug_auc.columns)
cellline_drug_auc_imputed.index = list(cellline_drug_auc.index)
cellline_drug_auc_imputed
```

```python
print("Mean", np.mean(cellline_drug_auc_imputed.values))
print("Min", np.min(cellline_drug_auc_imputed.values))
print("Max", np.max(cellline_drug_auc_imputed.values))
print("Median", np.median(cellline_drug_auc_imputed.values))
```

```python
audrc_csv = cellline_drug_auc
```

```python
audrc_csv_binary = pd.DataFrame(cellline_drug_auc_imputed.values < 0.25)
audrc_csv_binary = audrc_csv_binary.astype(int)
audrc_csv_binary.index = list(cellline_drug_auc.index)
audrc_csv_binary.columns = list(cellline_drug_auc.columns)
audrc_csv_binary
```

```python
audrc_cell_count = audrc_csv_binary.shape[0] * audrc_csv_binary.shape[1]
print(audrc_cell_count)
```

```python
print(f"Percentage of ones in AUDRC data = {np.count_nonzero(audrc_csv_binary.values)/audrc_cell_count}")
print(f"Percentage of zeros in AUDRC data = {1-(np.count_nonzero(audrc_csv_binary.values)/audrc_cell_count)}")
```
